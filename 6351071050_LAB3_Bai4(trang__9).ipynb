```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import warnings

# Bỏ qua tất cả các cảnh báo
warnings.filterwarnings('ignore')

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

print(f"Shape: {X.shape}, Classes: {target_names}\n")


max_values = X.max(axis=0)
X_normalized = X / max_values

print("Lý Thanh Nguyên - 6351071050")
```

câu c)

```python
def create_model():
    model = Sequential()
    # Hidden layer 1: 10 neurons, Relu
    model.add(Dense(10, activation='relu', input_shape=(4,)))
    # Hidden layer 2: 20 neurons, Relu
    model.add(Dense(20, activation='relu'))
    # Output layer: 3 classes, Softmax
    model.add(Dense(3, activation='softmax'))

    # Biên dịch mô hình
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
    print("Lý Thanh Nguyên - 6351071050")
```

câu d)

```python
k = 10
kfold = KFold(n_splits=k, shuffle=True, random_state=42)


models = []  # Lưu mô hình
fold_accuracies = []
fold_precisions = []
fold_recalls = []
fold_f1s = []


# Mỗi phần tử sẽ là một mảng [score_class_0, score_class_1, score_class_2]
class_precisions_list = []
class_recalls_list = []
class_f1s_list = []

fold_no = 1

for train_index, test_index in kfold.split(X_normalized):
    # Chia dữ liệu
    X_train, X_test = X_normalized[train_index], X_normalized[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Tạo model mới
    model = create_model()


    # verbose=0
    model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=0)

    # Dự đoán
    y_pred_probs = model.predict(X_test, verbose=0)
    y_pred = np.argmax(y_pred_probs, axis=1)

    # --- Tính toán các độ đo ---

    # 1. Accuracy
    acc = accuracy_score(y_test, y_pred)

    # 2. Precision, Recall, F1 (Average Macro - trung bình chung của fold)
    prec_macro, rec_macro, f1_macro, _ = precision_recall_fscore_support(
        y_test, y_pred, average='macro', zero_division=0
    )

    # 3. Precision, Recall, F1 (Per Class - cho từng lớp cụ thể)
    prec_class, rec_class, f1_class, _ = precision_recall_fscore_support(
        y_test, y_pred, average=None, zero_division=0, labels=[0, 1, 2]
    )

    # --- Lưu trữ kết quả ---
    models.append(model)

    fold_accuracies.append(acc)
    fold_precisions.append(prec_macro)
    fold_recalls.append(rec_macro)
    fold_f1s.append(f1_macro)

    class_precisions_list.append(prec_class)
    class_recalls_list.append(rec_class)
    class_f1s_list.append(f1_class)

    # In kết quả của từng fold
    print(f"Fold {fold_no}: Acc={acc:.4f}, Prec={prec_macro:.4f}, Rec={rec_macro:.4f}, F1={f1_macro:.4f}")

    fold_no += 1

print("\nĐã hoàn thành huấn luyện.\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu e)

```python
avg_acc = np.mean(fold_accuracies)
avg_prec = np.mean(fold_precisions)
avg_rec = np.mean(fold_recalls)
avg_f1 = np.mean(fold_f1s)

print(f"Average Accuracy:  {avg_acc:.4f}")
print(f"Average Precision: {avg_prec:.4f}")
print(f"Average Recall:    {avg_rec:.4f}")
print(f"Average F1-score:  {avg_f1:.4f}\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu f)

```python
avg_class_prec = np.mean(class_precisions_list, axis=0)
avg_class_rec = np.mean(class_recalls_list, axis=0)
avg_class_f1 = np.mean(class_f1s_list, axis=0)

df_class_metrics = pd.DataFrame({
    'Class': target_names,
    'Avg Precision': avg_class_prec,
    'Avg Recall': avg_class_rec,
    'Avg F1-score': avg_class_f1
})

print(df_class_metrics.to_string(index=False))
print("Lý Thanh Nguyên - 6351071050")
```

câu g)

```python
best_fold_index = np.argmax(fold_f1s)
best_model = models[best_fold_index]
print(f"Sử dụng mô hình từ Fold {best_fold_index + 1} (F1={fold_f1s[best_fold_index]:.4f})")

# Dữ liệu mới
new_samples = np.array([
    [6.2, 2.9, 4.3, 1.3],
    [5.1, 3.5, 1.4, 0.2],
    [7.3, 2.8, 6.4, 2.1]
])

# CHUẨN HÓA dữ liệu mới bằng max_values của tập gốc
new_samples_norm = new_samples / max_values

# Dự đoán
predictions_prob = best_model.predict(new_samples_norm, verbose=0)
predictions_class = np.argmax(predictions_prob, axis=1)

print("\nKẾT QUẢ DỰ ĐOÁN:")
for i, sample in enumerate(new_samples):
    class_name = target_names[predictions_class[i]]
    print(f"Mẫu {sample} -> Dự đoán: {class_name}")
    print("Lý Thanh Nguyên - 6351071050")
```
