```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
```

câu a)

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print("Lý Thanh Nguyên - 6351071050")

print("Đã tải xong dữ liệu MNIST.\n")
```

câu b)

```python
print(f"Train images shape: {train_images.shape}")
print(f"Train labels shape: {train_labels.shape}")
print(f"Test images shape:  {test_images.shape}")
print(f"Test labels shape:  {test_labels.shape}\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu c)

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
print("Lý Thanh Nguyên - 6351071050")
```

câu d)

```python
train_images, val_images, train_labels, val_labels = train_test_split(
    train_images, train_labels, test_size=0.2, random_state=42
)

print(f"Train (mới) shape:      {train_images.shape}")
print(f"Validation shape:       {val_images.shape}\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu e)

```python
model = Sequential()

# Flatten layer: Chuyển 2D (28x28) thành 1D (784)
model.add(Flatten(input_shape=(28, 28)))

# Dense layer: 128 neurons, activation Relu
model.add(Dense(128, activation='relu'))

# Dropout layer: 20%
model.add(Dropout(0.2))

# Dense layer (Output): 10 neurons (cho 10 số từ 0-9), activation Softmax
model.add(Dense(10, activation='softmax'))

model.summary()
print("Lý Thanh Nguyên - 6351071050")
```

câu f)

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
print("Đã compile model.\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu g)

```python
history = model.fit(train_images, train_labels,
                    epochs=5,
                    batch_size=32,
                    validation_data=(val_images, val_labels))
print("Lý Thanh Nguyên - 6351071050")
```

câu h)

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(12, 5))

# Biểu đồ Accuracy

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

# Biểu đồ Loss
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()
print("Lý Thanh Nguyên - 6351071050")
plt.show()
```

câu i)

```python
# predict trả về xác suất của 10 lớp
y_pred_probs = model.predict(test_images)
# argmax chọn ra lớp có xác suất cao nhất
y_pred = np.argmax(y_pred_probs, axis=1)
print("Lý Thanh Nguyên - 6351071050")
```

câu j)

```python
test_acc = accuracy_score(test_labels, y_pred)
print(f"\nAccuracy trên tập test: {test_acc:.4f}\n")
print("Lý Thanh Nguyên - 6351071050")
```

câu k)

```python
target_names = [str(i) for i in range(10)]
print(classification_report(test_labels, y_pred, target_names=target_names))
print("Lý Thanh Nguyên - 6351071050")
```

câu l)

```python
cm = confusion_matrix(test_labels, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix - MNIST')
plt.ylabel('Thực tế (True)')
plt.xlabel('Dự đoán (Predicted)')
plt.show()
print("Lý Thanh Nguyên - 6351071050")
```

câu m)

```python
plt.figure(figsize=(10, 2))
for i in range(5):
    # Lấy ảnh
    img = test_images[i]
    # Thêm chiều batch: (28, 28) -> (1, 28, 28) để đưa vào model
    img_expanded = np.expand_dims(img, axis=0)

    # Dự đoán
    pred_prob = model.predict(img_expanded, verbose=0)
    pred_label = np.argmax(pred_prob)
    true_label = test_labels[i]

    # Vẽ
    plt.subplot(1, 5, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(f"P:{pred_label} | T:{true_label}")
    plt.axis('off')
plt.show()
print("Lý Thanh Nguyên - 6351071050")
```

câu n)

```python
errors_indices = np.where(y_pred != test_labels)[0] # Lấy index các mẫu sai
plt.figure(figsize=(10, 2))

# Chỉ lấy 5 lỗi đầu tiên
for i, idx in enumerate(errors_indices[:5]):
    img = test_images[idx]
    pred_label = y_pred[idx]
    true_label = test_labels[idx]

    plt.subplot(1, 5, i+1)
    plt.imshow(img, cmap='gray')
    # Tiêu đề: P = Predicted (Dự đoán), T = True (Thực tế)
    plt.title(f"P:{pred_label} | T:{true_label}", color='red')
    plt.axis('off')
plt.show()
print("Lý Thanh Nguyên - 6351071050")
```
